```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import calendar

# Helper: load a store and return a pandas DataFrame indexed by (year, month) with column=member name
# This version filters years up to 2100 and applies unit conversion when appropriate.
def member_monthly_df(zstore, name, var='pr',year_max=2100):
    try:
        ds = xr.open_dataset(zstore, engine='zarr')
        da = ds[var]
        units = da.attrs.get('units', '')
        # determine spatial dims (everything except time)
        spatial_dims = [d for d in da.dims if d != 'time']
        if spatial_dims:
            da_mean = da.mean(dim=spatial_dims, skipna=True)
        else:
            da_mean = da

        # extract raw values and times
        times = da_mean['time'].values
        vals = da_mean.values

        # Apply unit conversion if needed (common for 'pr' in CMIP: kg m-2 s-1 -> mm/day)
        scale = 1.0
        out_units = units if units else ''
        if units and 'kg m-2 s-1' in units:
            scale = 86400.0  # seconds/day
            out_units = 'mm/day'
            vals = vals * scale
        elif units and ('m s-1' in units or 'm s-1' == units.strip()):
            # example fallback (not expected for pr) - convert m/s to mm/day
            scale = 864000.0
            out_units = 'mm/day'
            vals = vals * scale

        # times may be numpy.datetime64 or cftime objects; extract year/month safely
        years = []
        months = []
        for t in times:
            try:
                ts = pd.to_datetime(t)
                years.append(int(ts.year))
                months.append(int(ts.month))
            except Exception:
                years.append(int(getattr(t, 'year')))
                months.append(int(getattr(t, 'month')))

        dfm = pd.DataFrame({'year': years, 'month': months, name: vals})
        # Filter to years up to year_max (inclusive)
        dfm = dfm[dfm['year'] <= year_max]
        # Aggregate in case there are multiple entries per month
        dfm = dfm.groupby(['year', 'month'])[name].mean()
        dfm = dfm.to_frame()

        print(f"Loaded {name}: units='{units}' -> using scale={scale}, output_units='{out_units}', rows={len(dfm)}")
        return dfm, out_units
    except Exception as e:
        print(f"Error loading {zstore}: {e}")
        return None, None

# Collect per-member monthly DataFrames
member_dfs = []
member_units = []
for idx, row in df_filtered.iterrows():
    z = row['zstore']
    member_name = f"{row.get('experiment_id','')}_{row.get('member_id','')}_{idx}"
    print(f"Loading: {member_name} -> {z}")
    dfm, out_units = member_monthly_df(z, member_name, year_max=2100)
    if dfm is not None:
        member_dfs.append(dfm)
        member_units.append(out_units)

if len(member_dfs) == 0:
    raise RuntimeError('No monthly dataframes were loaded. Check zstore access and that xarray/zarr/fsspec (and gcsfs) are available.')

# Combine along columns (index is MultiIndex (year,month))
df_combined = pd.concat(member_dfs, axis=1)
# Compute ensemble mean across members
df_combined['ensemble_mean'] = df_combined.mean(axis=1)

# Turn index into columns
df_combined = df_combined.reset_index()
# Compute decade
df_combined['decade'] = (df_combined['year'] // 10) * 10

# Pivot to decade x month using ensemble_mean
pivot = df_combined.groupby(['decade', 'month'])['ensemble_mean'].mean().unstack(level='month')
# Ensure months 1..12 are present
pivot = pivot.reindex(columns=range(1,13))

# Determine units to label colorbar (majority vote)
units_label = None
if member_units:
    units_label = pd.Series(member_units).value_counts().idxmax()

# Print pivot stats for debug
print('Pivot stats â€” min:', pivot.min().min(), 'max:', pivot.max().max())

# Choose annotation format: if values small use scientific
maxval = pivot.max().max()
if pd.isna(maxval):
    raise RuntimeError('Pivot table is empty (NaNs). Check data selection and date range up to 2100.')

if abs(maxval) < 0.01:
    annot_fmt = '.4e'
else:
    annot_fmt = '.2f'

# Plot heatmap with month names on x-axis and reversed y-axis so earlier decades are at the bottom
plt.figure(figsize=(11, max(4, 0.6 * len(pivot))))
sns.set_theme(style='white')
month_names = [calendar.month_name[i] for i in range(1,13)]
ax = sns.heatmap(pivot, cmap='crest', annot=True, fmt=annot_fmt, cbar_kws={'label': f'pr ({units_label})'}, xticklabels=month_names)
ax.set_ylabel('Decade')
ax.set_xlabel('Month')
ax.set_title('Monthly Mean Precipitation Across Decades (2010 - 2100)')
# Reverse y-axis so earliest decade appears at the bottom and latest at the top
ax.invert_yaxis()
# Make sure y tick labels are integers (decade years)
ax.set_yticklabels([int(lbl.get_text()) for lbl in ax.get_yticklabels()])
# Rotate x labels for readability
ax.set_xticklabels(month_names, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Bar chart: precipitation variability (std dev) by decade for major monsoon regions
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import xarray as xr
import numpy as np

# Define monsoon regions by lat/lon bounds
monsoon_regions = {
    'Indian Summer Monsoon': {'lat': (5, 35), 'lon': (65, 100)},
    'West African Monsoon': {'lat': (5, 20), 'lon': (-20, 10)},
    'East Asian Monsoon': {'lat': (25, 50), 'lon': (100, 140)}
}

def compute_variability_by_region(zstore, regions_dict, var='pr', year_max=2100):
    """
    Load a zarr dataset and compute std dev of precipitation by (decade, region).
    Returns a DataFrame with columns: [decade, region, std_dev]
    """
    try:
        ds = xr.open_dataset(zstore, engine='zarr')
        da = ds[var]
        
        # Print available dimensions and coordinates for debugging
        print(f"  Dataset dims: {da.dims}")
        print(f"  Dataset coords: {list(da.coords.keys())}")
        
        # Get units and apply conversion
        units = da.attrs.get('units', '')
        scale = 1.0
        if units and 'kg m-2 s-1' in units:
            scale = 86400.0
            da = da * scale
        
        # Extract time coordinates
        times = da['time'].values
        years = []
        for t in times:
            try:
                ts = pd.to_datetime(t)
                years.append(int(ts.year))
            except Exception:
                years.append(int(getattr(t, 'year')))
        
        # Filter to year_max
        mask = np.array(years) <= year_max
        da_filtered = da.isel(time=mask)
        years_filt = np.array(years)[mask]
        
        # Compute variability (std dev) for each region
        results = []
        for region_name, bounds in regions_dict.items():
            lat_min, lat_max = bounds['lat']
            lon_min, lon_max = bounds['lon']
            
            try:
                # Try to select by lat/lon without method argument
                da_region = da_filtered.sel(lat=slice(lat_min, lat_max), lon=slice(lon_min, lon_max))
                
                # Compute spatial mean, then std dev across time
                spatial_mean = da_region.mean(dim=['lat', 'lon'], skipna=True)
                
                # Group by decade and compute std dev
                decades = (years_filt // 10) * 10
                for decade in sorted(np.unique(decades)):
                    decade_mask = decades == decade
                    if decade_mask.sum() > 0:
                        decade_vals = spatial_mean.values[decade_mask]
                        std_dev = float(np.nanstd(decade_vals))
                        results.append({'decade': int(decade), 'region': region_name, 'std_dev': std_dev})
                print(f"  âœ“ {region_name}: {len([r for r in results if r['region']==region_name])} decades")
            except Exception as e:
                print(f"  âœ— {region_name}: {e}")
        
        return results
    except Exception as e:
        print(f"Error computing variability: {e}")
        return []

# Process first member as an example
print("Computing variability by monsoon region...")
first_zstore = df_filtered['zstore'].iloc[0]
print(f"Using zstore: {first_zstore}")
var_results = compute_variability_by_region(first_zstore, monsoon_regions, year_max=2100)

if len(var_results) > 0:
    df_var = pd.DataFrame(var_results)
    print(f"Variability data: {len(df_var)} rows")
    
    # Plot bar chart
    plt.figure(figsize=(12, 6))
    sns.set_theme(style='whitegrid')
    sns.barplot(data=df_var, x='decade', y='std_dev', hue='region', palette='Set2')
    plt.xlabel('Decade')
    plt.ylabel('Std Dev (mm/day)')
    plt.title('Precipitation Variability by Decade and Monsoon Region')
    plt.legend(title='Monsoon Region', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
else:
    print("No variability data computed. Check zstore coordinates and region bounds.")
```
